```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/01_raw/sessions.csv")
print(df.info())
print(df.describe())
```

```python
# Target distribution
sns.countplot(x="ordered", data=df)
plt.title("Distribution of Target (ordered)")
plt.show()
```

```python
# Feature correlations
plt.figure(figsize=(12,10))
sns.heatmap(df.drop(columns=['UserID']).corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Grouped feature means
means = df.groupby("ordered").mean().T
means.plot(kind="bar", figsize=(15,7))
plt.title("Mean of Features Grouped by Target")
plt.tight_layout()
plt.show()
```
